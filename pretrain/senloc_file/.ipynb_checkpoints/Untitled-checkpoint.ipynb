```python
all_channels = mi_channels+large_im_channels+bci_comp_iv2a_channels+bci_comp_iv2b_channels+bci_meditation_channels+hgd_channels
```

```python
unique_channels= set(all_channels)
len(unique_channels)
```

```python
channel_to_indice_dict = {}
chan_ind = 0
for chan in unique_channels:
    channel_to_indice_dict[chan] = chan_ind
    chan_ind+=1
```

```python
mi_channel_ind = np.array([channel_to_indice_dict[p] for p in mi_channels])
large_im_channel_ind = np.array([channel_to_indice_dict[p] for p in large_im_channels])
bci_iv2a_channel_ind = np.array([channel_to_indice_dict[p] for p in bci_comp_iv2a_channels])
bci_comp_iv2b_channel_ind = np.array([channel_to_indice_dict[p] for p in bci_comp_iv2b_channels])
bci_meditation_channel_ind = np.array([channel_to_indice_dict[p] for p in bci_meditation_channels])
hgd_channel_ind = np.array([channel_to_indice_dict[p] for p in hgd_channels])
```

```python
channel_dict = {}
channel_dict["channels_mapping"] = channel_to_indice_dict
channel_dict["bci_comp_iv2a"] = bci_iv2a_channel_ind
channel_dict["bci_comp_iv2b"] = bci_comp_iv2b_channel_ind
channel_dict["eeg_mi_bci"] = mi_channel_ind
channel_dict["large_im"] = large_im_channel_ind
channel_dict["bci_meditation"] = bci_meditation_channel_ind
channel_dict["hgd"] = hgd_channel_ind
```

```python
with open("/vsc-hard-mounts/leuven-data/343/vsc34340/EEG_ITransformer/senloc_file/sen_chan_idx.pkl", 'wb') as file:
    pickle.dump(channel_dict, file)
```
